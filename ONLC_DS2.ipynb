Lets use cross-validation to measure a model's effectiveness.  Treat a 10th of the data as Test, the rest as Training, and score effectiveness on the remaining 10th after each training.

* [Decision Tree Classifier](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

* [Cross Validation](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html)

In this notebook, I highlight your ability to embed Youtubes, by selecting three from one of my favorite on-line Machine Learning teachers.

```python
from IPython.display import YouTubeVideo
YouTubeVideo("DCZ3tsQIoGU")
```

```python
>>> from sklearn.datasets import load_iris
>>> from sklearn.model_selection import cross_val_score
>>> from sklearn.tree import DecisionTreeClassifier
>>> clf = DecisionTreeClassifier(random_state=0)
>>> iris = load_iris()
>>> cross_val_score(clf, iris.data, iris.target, cv=10)
```

[Random Forest Classifier](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html)

```python
YouTubeVideo("D_2LkhMJcfY")
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0)
cross_val_score(clf, iris.data, iris.target, cv=10)
```

[Support Vector Classifier](http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html)

```python
YouTubeVideo("Y6RRHw9uN9o")
```

```python
>>> from sklearn.svm import SVC
>>> clf = SVC()
```

```python
cross_val_score(clf, iris.data, iris.target, cv=10)
```
