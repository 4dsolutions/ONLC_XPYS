```python
domain = np.linspace(-3, 3, 100)
range1 = np.cos(domain)
range2 = np.sin(domain)
plt.plot(domain,range1,"r", domain,range2,"k+");
```

```python
plt.figure(0)
plt.xlabel("Domain")
plt.ylabel("Range")
plt.title("Trig Homework")
plt.plot(domain, range1, color="cyan", label="Cosine")
plt.plot(domain, range2, color="red", linestyle="dashed", label="Sine")
plt.legend();
```

```python
data=[2,8,3,6,1]
xaxis=[1,2,3,4,5]
plt.bar(xaxis, height=data, ec=("r","b","g","y","k"), lw = 3)
plt.annotate('Tall!', xy=[1.5,6], xytext=[0.5,5.9], 
             arrowprops={'facecolor':'black', 'shrink':0.05})
plt.annotate('Small!', xytext=[4.7,2], xy=[4.9,1.1], 
             arrowprops={'facecolor':'red', 'shrink':0.05});
```

Cutting and pasting from documentation, so to better understand:

[hexbin](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.hexbin.html)

```python
>>> n = 10000
>>> df = pd.DataFrame({'x': np.random.randn(n),
...                    'y': np.random.randn(n)})  # dict of 2 series
>>> df.plot.hexbin(x='x', y='y', gridsize=20);
```

```python
>>> n = 500
>>> np.random.seed(42)
>>> df = pd.DataFrame({
...     'coord_x': np.random.uniform(-3, 3, size=n),
...     'coord_y': np.random.uniform(30, 50, size=n),
...     'observations': np.random.randint(1,5, size=n)
...     })
>>> ax = df.plot.hexbin(x='coord_x',
...                     y='coord_y',
...                     C='observations',
...                     reduce_C_function=np.mean,
...                     gridsize=10,
...                     cmap="viridis")
```
